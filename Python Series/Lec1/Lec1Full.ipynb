4. Identity Operator

  The 'is' operator returns True only if two variables point to the same object in memory. The 'is not' returns the inverse.

````python
a = ["Orange"]
b = ["Orange"]
c = a

print(a is b)

print(a is c)

print(a is not b)

print(a == b)
````


'is' Checks if both variables point to the same object in memory

'==' Checks if the values of both variables are equal

Therefore, in the above example, while the value of a and b is same, they correspond to two different objects. Hence, statement 1 is False.

However, since a is assigned to c, they share a common object (in the memory).
Which is why statement 2 returns True.

### Range

The command range() is used to generate a sequence of numbers. It is most commonly used in loops, to control how many times a block of code executes. The range function is efficient because it generates numbers on demand rather than storing them all in memory at once.

````python
i = list(range(5)) # the 'list' command here is used to store multiple items in a single variable.
print(i)
````

The command may take 1, 2 or 3 parameters.
```
range(start, stop, step)
```
However, the start and the step are optional. By default, the start is set at 0 and the step is set as 1. The start is inclusive, while the stop is not. Hence, the above example can be thought as equivalent to:

```
range(0, 5, 1)
```

### Loops
Python has two primitive loop commands:

   `while`
    `for`


The 'while' loop executes a code block as long as a condition is True.

````python
# 1. Initialization: set up a counter variable
counter = 0

# 2. Condition: the while loop keeps running as long as counter < 5
while counter < 5:
    # 3. Body: code inside the loop, will repeat as long as condition is True
    print("Counter is now:", counter)

    # 4. Update: advance the counter so that the loop can end eventually
    counter = counter + 1  # Alternatively: counter += 1

# Loop ends automatically when counter reaches 5 (and therefore counter < 5 is now False)
print("Loop finished! Final value of counter:", counter)
````

What happens when the condition is always true?

````python
counter=2

while counter > 1:
    print("Counter is now:", counter)

    counter +=1
print("Loop finished! Final value of counter:", counter)
````

The 'for' loop is used for iterating over a sequence of objects.

````python

# 1. Initialization/Setup: create a sequence to iterate through
numbers = [0, 1, 2, 3, 4]

# 2. Iteration: for each value in 'numbers', assign it to 'num'
for num in numbers:
    # 3. Body: this code runs on each value of 'num'
    print("The current number is:", num)

    # 4. Update: happens automatically, 'num' takes the next value in the next loop

# Loop ends automatically after the last element
print("Loop finished!")
````

````python
a = 1
b = 1
print(a)
print(b)
for i in range(10):
    temp = a + b
    a = b
    b = temp
    print(temp)
````

Some other elements of loops are:
   `break`
    `continue`

The 'break' statement stops the loop before it has looped through all the items:

````python
fruit = 'Strawberry'
for x in fruit:
  print(x)
  if x == 'b':
    break
````

Notice the difference in output depending on the position of the 'break' condition.

````python
fruit = 'Strawberry'
for x in fruit:
  if x == 'b':
    break
  print(x)
````

With the 'continue' statement, the current iteration of the loop can be skipped and continued with the next one:

````python
fruit = 'Strawberry'
for x in fruit:
  if x == 'b':
    continue
  print(x)    #This line will not execute when x==b
````
