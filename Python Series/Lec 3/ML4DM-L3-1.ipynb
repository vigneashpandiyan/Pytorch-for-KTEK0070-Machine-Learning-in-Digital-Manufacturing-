### Lists
Lists can hold multiple datatypes and even duplicates. A loop may be used to
extract elements one at a time.

````python
list = [33, "hello", 3.5, True, 33]
print(list)
````

````python
names = ["Anna", "Bob", "Catherine"]

for name in names:
    print(f"{name} has {len(name)} letters.")
````

List items are indexed (ordered): the first item has index [0], the second item has index [1] etc. Therefore, it is possible to retrieve item for a specific index or even replace it. A negative indice starts from the end of the list and counts backwards towards the start.

````python
fruits = ["apple", "banana", "cherry"]
print(fruits)           # The original list
print(fruits[1])        # The second list item
print(fruits[-1])       # The last list item

fruits[1] = "blueberry" # Change value of second list item
print(fruits)           # The modified list
````

Some simple operations regarding lists are:

Additive

    Use append() to add at the end.
    Use insert() to add at index position.
    Use extend() to add multiple elements.

Substractive


    Use remove() to delete by value
    Use pop() to remove by index.
    Use clear() to empty the list.


````python
fruits = ["apple", "banana", "cherry", "date"]  #To reset the list, execute this block again.
````

````python
fruits.append("elderberry") #Frequently used in loops
print(fruits)
fruits.insert(1, "fig") #First parameter is the position, second parameter is the item to be inserted
print(fruits)
fruits.extend(["grape", "honeydew"])
print(fruits)
````

````python
fruits.remove("cherry") #The first occurance only
print(fruits)
fruits.pop(2)
print(fruits)
fruits.clear()
print(fruits)
````

The sort() method sorts the list ascending by default. The optional 'reverse' parameter can be set to True for descending order.

````python
numbers = [5, 2, 9, 1]
print("Original list",numbers)
numbers.sort()
print("Sorted ascending", numbers)

numbers.sort(reverse=True)
print("Sorted descending", numbers)
````

Note: The list maybe sorted in place like in the example above (i.e. original list is modified) or the sorted list maybe another variable like below ( original list is unmodified):

### List of List

List may include lists as items. This is an important functionality which enables us to create primitive equivalent of matrices/arrays in python.

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix)
print(matrix[2][2]) #Reminder that indices start from 0, therefore, here the third item of the third list is queried
````

The len() function returns the number of items in an object.

When used on the matrix, it will return the number of rows; when used on the row, it will return the number of columns.

````python
A = [
    [1, 2],
    [3, 4]
]

print(len(A))

print(len(A[0]))
````

````python
B = [
    [5],
    [6]
]

print(len(B))

print(len(B[0]))
````

A rudimentary matrix multiplication is presented below.
Note the use of 'for' loops, range, append and  len() functions.

````python
result = []
for i in range(len(A)):
    row = []
    for j in range(len(B[0])):
        sum = 0
        for k in range(len(B)):
            sum += A[i][k] * B[k][j]
        row.append(sum)
    result.append(row)

print("Matrix Multiplication:\n", result)
````

Exercise:

Given the following square matrix $A$
$$A =  \begin{pmatrix} 1 & 1 \\ 1 & 0  \end{pmatrix}$$
compute the matrix $A^n$ obtained by multiplying $A$ to itself $n$ times.

Can you comment on the left-top value in $A^n$ (what does this remind you of?)

````python
def square_matrix_power(A, n):
    output = A
    temp = A
    for i in range(n - 1):
        # output = A * output
        # -> output = A * temp
        #    temp = output
        output[0][0] = A[0][0] * temp[0][0] + A[0][1] * temp[1][0]
        output[0][1] = A[0][0] * temp[0][1] + A[0][1] * temp[1][1]
        output[1][0] = A[1][0] * temp[0][0] + A[1][1] * temp[1][0]
        output[1][1] = A[1][0] * temp[0][1] + A[1][1] * temp[1][1]
        temp = output
    return output
````

````python
A = [
     [1, 1],
     [1, 0]
]
````

````python
print(square_matrix_power(A, 2))
````

````python
def square_matrix_power(A, n):
    output = [item[:] for item in A]
    temp = [item[:] for item in A]
    for i in range(n - 1):
        # output = A * output
        # -> output = A * temp
        #    temp = output
        output[0][0] = A[0][0] * temp[0][0] + A[0][1] * temp[1][0]
        output[0][1] = A[0][0] * temp[0][1] + A[0][1] * temp[1][1]
        output[1][0] = A[1][0] * temp[0][0] + A[1][1] * temp[1][0]
        output[1][1] = A[1][0] * temp[0][1] + A[1][1] * temp[1][1]
        temp = [item[:] for item in output]
    return output
````

````python
A = [
     [1, 1],
     [1, 0]
]
````

````python
print(square_matrix_power(A, 2))
````

````python
for n in range(10):
    output = square_matrix_power(A, n)
    print(n, output[0][0])
````

### Dictionary

The last variation of lists is dictionaries. They store data in key:value pairs.

A dictionary is a collection which is ordered, changeable and does not allow duplicates.

````python
car =	{
  "brand": "Dodge",
  "model": "Charger",
  "year": 1969
}
print(car)
print(car["brand"])
````

Since duplicate keys are not allowed, the next occurance of same key will overide the previous.

````python
car =	{
  "brand": "Dodge",
  "model": "Charger",
  "year": 1969,
  "year": 2015
}
print(car)
````
