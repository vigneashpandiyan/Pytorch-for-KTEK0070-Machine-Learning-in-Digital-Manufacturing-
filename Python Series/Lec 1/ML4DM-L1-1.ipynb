What is the Colab Runtime?

When you use Google Colab, your code doesn't run directly on your computer. Instead, it runs on a virtual computer (or "runtime") hosted in Google's cloud. This is called the Colab runtime.

Colab runtimes are temporary. Sessions can last up to 12 hours but may be disconnected sooner if idle.
Files and variables created in the runtime are **lost** when the session ends (unless saved to Google Drive).


You can always check the current session status at the top-right of the Colab window, where a green tick (or a warning symbol) indicates if your runtime is connected.

Note: Google Colab is based on Jupyter Notebooks.

### Comments

In Python, comments start with the # symbol. Anything after # on a line is ignored by the Python interpreter.

````python
# This is a comment explaining what the next line does
x = 5  # Set x to 5

# Now let's print x
print("The value of x is:", x)
````

The comments (# ...) are there to help humans understand the code.
