6.   Choose two real numbers $p$ and $q$ and compute the following two values
$$ v_1 = F(p, 100) * F(q, 100) $$
$$ v_2 = F(p + q, 100) $$
Compute the difference $v_1 - v_2$ and comment on what you see.

# Solution to Problem 2

```python
def factorial(n):
    product = 1
    for i in range(n):
        # print(i, product)
        product *= (i + 1)
    return product
```

```python
print(factorial(3))
```

```python
def factorial_recursive(n):
    if n > 1:
        return n * factorial_recursive(n - 1)
    else:
        return 1
```

```python
print(factorial_recursive(3))
```

```python
def compute_ratio(x, n):
    ratio = x**n / factorial(n)
    return ratio
```

```python
compute_ratio(2, 3)
```

```python
def compute_sum(x, N):
    sum = 1
    for i in range(N):
        sum += compute_ratio(x, i + 1)
    return sum
```

```python
print(compute_sum(2, 3))
```

```python
def compute_sum_epsilon(x, epsilon):
    sum = 1
    var = epsilon
    i = 1
    while var >= epsilon:
        var = compute_ratio(x, i)
        sum += var
        i += 1
    return sum
```

```python
print(compute_sum_epsilon(2, 0.01))
```

```python
print(compute_sum(2, 100))
```

```python
def compute_sum_epsilon(x, epsilon):
    sum = 1
    i = 1
    while True: 
        var = compute_ratio(x, i)
        sum += var
        i += 1
        if var < epsilon:
            break
    return sum
```

```python
print(compute_sum_epsilon(2, 0.01))
```

```python
p = -1.5
q = 7.1

v1 = compute_sum(p, 100) * compute_sum(q, 100)
v2 = compute_sum(p + q, 100)

print(v1, v2, v2 - v1)
```
