More dataframes methods:

````python
def create_df(nRows, nCols, maxRand=10):
    arr = np.random.randint(0, maxRand, (nRows, nCols))
    df = pd.DataFrame(arr)
    df.index = ['R' + str(x) for x in np.arange(1, nRows+1)]
    df.columns = ['C' + str(x) for x in np.arange(1, nCols+1)]
    return df
````

````python
create_df(5, 3)
````

````python
create_df(2, 5)
````

A csv (comma separated values) file may be directly loaded into a DataFrame.

````python
import pandas as pd
cht = pd.read_csv('/content/sample_data/california_housing_test.csv') #Available in colab

print(cht)
````

By default, for large data, only a preview of the table is returned.

However,
```
to_string()
```
may be used to print the entire DataFrame.

````python
print(cht.to_string())
````

However, it is usually more convenient to limit the number of rows in the output. The head() method returns the headers and the specified number of rows, starting from the top. Meanwhile, the tail() method returns the headers and the specified number of rows, starting from the bottom.

````python
print(cht.head(10))
````

````python
print(cht.tail(10))
````

````python
 print(cht.info())
````

The above info() method gives more information about the data set including the number of rows (or observations), columns, datatypes of each column and the presence of any missing (null) values. In this case, there is no missing data, i.e. all 3000 observations contain non-null data.
